# Pandas
Pandas is a powerful Python library primarily used for data manipulation and analysis, offering efficient tools for working with structured data like spreadsheets and databases.

# Pandas Series:
A pandas Series is a one-dimensional labeled array in Python that can hold data of any type (integers, floats, strings, etc.).

It is  similar to a column in a spreadsheet or a single column in a DataFrame.

```python
# Importing libraries
import pandas as pd
import numpy as np
```

## Series from list!


```python
# Creating series from strings
country = ['India','Pakistan','USA','Srilanka']
pd.Series(country)
```

```python
# Creating series from integers
runs = [31,24,78,90,122]
pd.Series(runs)
```

```python
# Customizing index
marks = [67,58,68,85]
subject = ['english','maths','science','french']
pd.Series(marks,index = subject)
```

```python
# Naming your series object!
pd.Series(marks,index = subject, name = 'My marks!')
```

```python
# Storing it in a variable:
marks = pd.Series(marks,index = subject, name = 'My marks!')
marks
```

## Series from dictionaries


```python
# Initializing dict
marks = {
    'maths': 67,
    'english': 88,
    'french': 52,
    'science':87
}
```

```python
# Dict to series
marks_series = pd.Series(marks,name = 'My marks!')
```

```python
marks_series
```

## Series Attributes


```python
# size : tells the number of items in a series
marks_series.size
```

```python
# dtype : tells the data type of series elements
marks_series.dtype
```

```python
# name : tells the name of series
marks_series.name
```

```python
# is_unique : tells us if there are any unique elements or not
marks_series.is_unique
```

```python
# index : returns an index object containing names from index
marks_series.index
```

```python
# values : returns all the values from series
marks_series.values
```

```python
type(marks_series.values)
```
