# API
An API (Application Programming Interface) is a set of rules and tools that allows different software applications to communicate with each other. It acts as a bridge between different systems, enabling them to send and receive data in a structured way.

```python
import pandas as pd
import requests
```

The requests.get() function sends an HTTP GET request to the specified URL.

so basically making an HTTP request to fetch Top Rated Movies from TMDb.


```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=1be92787a3e92d93fbd228930580606c')
```

```python
# Create an Empty data frame
df = pd.DataFrame()
```

```python
for i in range(1, 500):  # Loop through pages 1 to 503
    response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=1be92787a3e92d93fbd228930580606c&page={i}")

    if response.status_code != 200:  # Check for errors
        print(f"Error fetching page {i}: {response.status_code}")
        break  # Stop if there's an error

    temp_df = pd.DataFrame(response.json().get('results', []),
                           columns=['id', 'title', 'release_date', 'overview', 'popularity', 'vote_average', 'vote_count'])

    df = pd.concat([df, temp_df], ignore_index=True)
```
