## 1. Scatterplot(Numerical-Numerical)

```python
sns.scatterplot(x='total_bill',y='tip',data = tips)
```

```python
# You can do analysis on 3 variables using scatterplot;Multi-variate analysis

sns.scatterplot(x='total_bill', y='tip',data=tips,hue ='sex' )
```

```python
# You can do analysis on 4 variables using scatterplot;Multi-variate analysis
sns.scatterplot(x='total_bill', y='tip',data=tips,hue ='sex' ,style ='smoker',size='size')
```

## 2.Bar Plot (Numerical-Categorical)

```python
titanic.head()
```

```python
titanic.info()
```

```python
# Pclass is a numerical Value convert it into string
titanic['Pclass'] = titanic['Pclass'].astype(str)
```

```python
# Age is a object convert into integer but first drop the NaN values that throw an error as they cannot be cverted into int

# Replace non-numeric values with NaN
titanic['Age'] = pd.to_numeric(titanic['Age'], errors='coerce')

# Now convert to integer (after dropping NaNs)
titanic['Age'] = titanic['Age'].dropna().astype(int)
```

```python
sns.barplot(x='Pclass',y='Age',data=titanic,hue='Sex')
```

## 3.Box Plot (Numerical-Categorical)

```python
# Sex is a object convert into integer but first drop the NaN values that throw an error as they cannot be cverted into int

# Replace non-numeric values with NaN
titanic['Sex'] = pd.to_numeric(titanic['Sex'], errors='coerce')

# Now convert to integer (after dropping NaNs)
titanic['Sex'] = titanic['Sex'].dropna().astype(int)
```

```python
import pandas as pd

titanic = pd.read_csv("titanic.tsv",sep='\t', na_values=["\\N"])
print(titanic[['Sex', 'Age']].info())  # Check if Sex is fixed
```
